```python
# Load the data
product_df = pd.read_csv('data/product_info_cleaned.csv')
product_df.head()
```

```python
# Reference: https://www.kaggle.com/code/themeeemul/sephora-eda-and-sentiment-analysis-using-pytorch
review_df_1 = pd.read_csv("data/reviews_0-250.csv",index_col = 0, dtype={'author_id':'str'})
review_df_2 = pd.read_csv("data/reviews_250-500.csv",index_col = 0, dtype={'author_id':'str'})
review_df_3 = pd.read_csv("data/reviews_500-750.csv",index_col = 0, dtype={'author_id':'str'})
review_df_4 = pd.read_csv("data/reviews_750-1250.csv",index_col = 0, dtype={'author_id':'str'})
review_df_5 = pd.read_csv("data/reviews_1250-end.csv",index_col = 0, dtype={'author_id':'str'})

# Merge review_df_1 till review_df_6
review_df = pd.concat([review_df_1, review_df_2, review_df_3, review_df_4, review_df_5],axis=0)
print("review_df shape: ",review_df.shape)
```

```python
columns = product_df.columns
filtered_columns = [col for col in columns if 'contains_' not in col and \
    'child_' not in col and 'price_per_' not in col and 'variation_' not in col and \
    'size_' not in col]
print(filtered_columns)
product_df = product_df[filtered_columns]
product_df.head()
```
